#Architecture

```python
import os
import matplotlib.pyplot as plt
from PIL import Image

def calculate_area_ratio(width, height, img_width, img_height):
    # Calculate the area ratio as a percentage of the image area
    bounding_box_area = width * height
    image_area = img_width * img_height
    area_ratio_percentage = (bounding_box_area / image_area) * 100
    return area_ratio_percentage

def find_image_file(image_folder_path, base_filename):
    possible_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff']
    for ext in possible_extensions:
        image_path = os.path.join(image_folder_path, base_filename + ext)
        if os.path.exists(image_path):
            return image_path
    return None

def process_annotation_file(file_path, img_width, img_height):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        area_ratios = []
        for line in lines:
            parts = line.strip().split()
            class_id = int(parts[0])
            center_x = float(parts[1])
            center_y = float(parts[2])
            width = float(parts[3]) * img_width
            height = float(parts[4]) * img_height
            area_ratio = calculate_area_ratio(width, height, img_width, img_height)
            area_ratios.append(area_ratio)
        return area_ratios

def process_annotations_folder(label_folder_path, img_folder_path):
    all_area_ratios = []
    for filename in os.listdir(label_folder_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(label_folder_path, filename)
            base_filename = os.path.splitext(filename)[0]
            image_path = find_image_file(img_folder_path, base_filename)
            if image_path:
                with Image.open(image_path) as img:
                    img_width, img_height = img.size
                area_ratios = process_annotation_file(file_path, img_width, img_height)
                all_area_ratios.extend(area_ratios)
    return all_area_ratios

def plot_histogram(area_ratios, title):
    plt.hist(area_ratios, bins=50, alpha=0.75)
    plt.xlabel('Area Ratio (%)')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Bounding Box Area Ratios - {title}')
    plt.grid(True)
    plt.show()
```

##Hist plot of fruit_fr

```python
# Define folder paths
base_folder_path = '/content/gdrive/MyDrive/yolov8_KLTN/datasets/fruit_fr/'
label_base_path = os.path.join(base_folder_path, 'labels')
img_base_path = os.path.join(base_folder_path, 'images')

# Define subfolders for train, val, test
subfolders = ['train', 'val', 'test']

for subfolder in subfolders:
    label_folder_path = os.path.join(label_base_path, subfolder)
    img_folder_path = os.path.join(img_base_path, subfolder)
    area_ratios = process_annotations_folder(label_folder_path, img_folder_path)

    if area_ratios:
        plot_histogram(area_ratios, subfolder.capitalize())
```

```python
import base64
import io
from PIL import Image
import anvil.server
import cv2

@anvil.server.callable
def predict_image(img_base64):
    # Decode the base64 string to bytes
    img_bytes = base64.b64decode(img_base64)

    # Convert bytes to a PIL Image
    img = Image.open(io.BytesIO(img_bytes))

    # Predict using YOLOv8
    results = model.predict(img)

    # Draw bounding boxes on the image
    result_img = results[0].plot()  # This is a numpy.ndarray

    # Convert numpy array to PIL Image
    # Make sure to convert BGR to RGB if necessary
    result_img_rgb = cv2.cvtColor(result_img, cv2.COLOR_BGR2RGB)
    result_img_pil = Image.fromarray(result_img_rgb)

    # Convert the PIL Image to bytes
    img_byte_arr = io.BytesIO()
    result_img_pil.save(img_byte_arr, format='PNG')
    img_bytes = img_byte_arr.getvalue()

    # Encode the bytes to a base64 string
    img_base64 = base64.b64encode(img_bytes).decode('utf-8')

    return img_base64
```
